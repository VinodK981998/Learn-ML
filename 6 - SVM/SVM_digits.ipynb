```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
digits = load_digits()
dir(digits)
```

```python
df = pd.DataFrame(digits.data,columns=digits.feature_names)
```

```python
inputs = digits.data
target = digits.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
model = SVC()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
param_grid = [{'C': [0.1, 1, 10, 100], 'gamma': [0.001, 0.01, 0.1, 1] },
              {'C': [0.1, 1, 10, 100], 'gamma': [0.001, 0.01, 0.1, 1], 'kernel' : ['rbf']},
              {'C': [0.1, 1, 10, 100], 'gamma': [0.001, 0.01, 0.1, 1], 'kernel' : ['linear']}]

# Create the SVC model
model = SVC()

# Create the GridSearchCV object
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1)

# Fit the GridSearchCV on the training data
grid_search.fit(X_train, y_train)

# Get the best parameters and best score from the cross-validation
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Get the best model based on cross-validation results
best_model = grid_search.best_estimator_

# Calculate the accuracy on the test set using the best model
test_accuracy = best_model.score(X_test, y_test)

# Create a DataFrame to save the results
results_df = pd.DataFrame(grid_search.cv_results_)



print("Best Parameters:", best_params)
print("Best Cross-Validation Score:", best_score)
print("Test Accuracy:", test_accuracy)
```
